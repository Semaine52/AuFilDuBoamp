# Constatations

#### Qui ne s'est jamais demandé ce qu'il s'est passé à Toulouse au BOAMP en 2018?   
#### Pour commencer, vous récupérez toute la colonne «ville» extraite des annonces de l'année.
#### Cela donne ceci:

```python
# IMPORTS DES OUTILS
from urllib.request import urlretrieve
import pandas as pd

# RÉGLAGE IPYTHON INTERACTVESHELL
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all" 

# RÉGLAGES PANDAS
pd.set_option('display.notebook_repr_html', True)
pd.set_option('display.max_colwidth', -1)
pd.set_option('display.max_rows', 2000)
pd.set_option('display.max_columns', 500)

# IMPORT ET LECTURE DES DONNÉES
URL = 'https://www.semaine52.com/AuFilDuBoamp/PKL/df_d01_08_ville_2018_no_index.pkl'
urlretrieve(URL, 'df_d01_08_ville_2018_no_index.pkl');
df = pd.read_pickle('df_d01_08_ville_2018_no_index.pkl')
a, _ = df.shape

print('*' * 100)
print(f"Nombre d'annonce au BOAMP en 2018: {a} ")
print(f"Ci-dessous, les cinq premières lignes du tableau de résultats:")

df.head()
```

____

#### Toujours en confiance, à partir des 115 935 entrées de départ, vous réduisez votre recherche à Toulouse:

```python
df_toulouse = df.loc[lambda df: df.d01_08_ville == 'Toulouse', :].reset_index(drop=True)
df_toulouse.shape
df_toulouse.head()
```

#### Vous auriez donc 795 annonces sur l'année 2018 à Toulouse en stock.
#### Mais il y a un problème, on le voit plus haut, c'est le fameux «cedex».

#### Je vous épargne ici les laborieuses étapes de nettoyage (qui feront l'objet d'un autre article), pour vous montrer ceci:



```python
df_boamp_2018_classe_par_nom_de_ville = df.sort_values(by='d01_08_ville').reset_index(drop=True)
df_boamp_2018_classe_par_nom_de_ville.iloc[0:10]
```

#### Ce qu'on y voit, c'est l'implémentation par l'acheteur de la colonne «ville» quand on classe la liste par ordre alphabétique.
#### Si vous lisez cet article sur JupyterLab (voir plus loin), je vous invite à vous balader un peu plus loin dans la base, c'est parfois cocasse...
